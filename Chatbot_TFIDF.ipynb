```python
import numpy as np
import nltk
import random
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import pandas as pd
import random

import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import warnings
warnings.filterwarnings("ignore")
```

```python
word_token = []

for col in df_objek.columns:
  get_token = word_tokenize(col)
  word_token += get_token
word_token
```

```python
stopword = nltk.download('stopwords')
stopwords_ind = stopwords.words('indonesian')
```

```python
def response(user_response):
  chatbot_response= ''
  sent_token.append(user_response)
  vec = TfidfVectorizer(stop_words=stopwords_ind)
  tfidf = vec.fit_transform(sent_token)
  vals = cosine_similarity(tfidf[-1], tfidf)
  flat = vals.flatten()[:-1]
  idx = flat.argmax()

  if(flat.max() == 0):
    print("Maaf kami tidak mengerti pertanyaan anda!")
  else:
    idx = flat.argmax()
    for i in range (0, len(df_objek[sent_token[idx]])):
      if (df_objek.at[i, sent_token[idx]] is not None):
        print(df_objek[sent_token[idx]].iloc[i])
      else:
        break
    sent_token.remove(sent_token[-1])
```

```python
user_intro = ['hai', 'halo', 'tes', 'pagi', 'siang', 'malam', 'hi']
bot_intro = ['Halo', 'Selamat Datang', 'Hai!', 'Senang berbicara dengan Anda!']

def greetings(text):
  for word in text.split():
    if word.lower() in user_intro:
      return random.choice(bot_intro)
```

```python
user_end = ['bye', 'exit', 'selesai', 'dadah', 'selamat tinggal']
bot_end = "Terima kasih telah menggunakan chatbot ini, selamat tinggal!"

def exit_bot(text):
  for word in text.split():
    if word.lower() in user_end:
      return bot_end
```

```python
flag = True

while flag:
  user_input = input()
  user_input = user_input.lower()
  if exit_bot(user_input) is not None:
    print(exit_bot(user_input))
    flag = False
  else:
    if greetings(user_input) is not None:
      print(greetings(user_input))
    else:
      response(user_input)
```
